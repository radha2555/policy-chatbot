```python
import torch
import gradio as gr
from transformers import pipeline

# Load the question-answering model
pipe = pipeline("question-answering", model="google-bert/bert-large-uncased-whole-word-masking-finetuned-squad")

def read_file_content(file_obj):
    """
    Reads the content of a file object and returns it.
    """
    try:
        with open(file_obj.name, 'r', encoding='utf-8') as file:
            return file.read()
    except Exception as e:
        return f"An error occurred: {e}"

def get_answer(file, prompt_type):
    """
    Extracts an answer based on the selected short prompt.
    """
    context = read_file_content(file)

    prompts = {
        "Loan Tenure": "What is the maximum tenure for a gold loan?",
        "Eligibility": "Who can apply for a gold loan?",
        "Loan Purpose": "What are the permitted uses of a gold loan?",
        "Minimum Loan Amount": "What is the minimum amount I can borrow?",
        "Gold Purity Requirement": "What is the minimum purity of gold required for a loan?",
        "Interest Charges": "How is interest calculated on gold loans?",
        "Pledged Gold Calculation": "How is the loan amount determined based on gold weight and purity?",
        "Interest Revision Policy": "How often are interest rates revised?"
    }

    question = prompts.get(prompt_type, "Provide details about gold loan policy.")
    answer = pipe(question=question, context=context)
    return answer["answer"]

demo = gr.Interface(
    fn=get_answer,
    inputs=[
        gr.File(label="Upload your file"),
        gr.Dropdown(
            label="Select a Question",
            choices=[
                "Loan Tenure", "Eligibility", "Loan Purpose", "Minimum Loan Amount",
                "Gold Purity Requirement", "Interest Charges", "Pledged Gold Calculation",
                "Interest Revision Policy"
            ]
        )
    ],
    outputs=gr.Textbox(label="Answer"),
    title="Gold Loan Policy QA",
    description="Upload a document and select a question to get an answer based on the document's content."
)

demo.launch(share=True)
```
